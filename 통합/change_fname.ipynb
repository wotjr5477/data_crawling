```python
import os
```

```python
fpath = os.getcwd()
fpath
```

```python
fnames = os.listdir(fpath)
fnames
```

```python
fnames = os.listdir(fpath)
end = r".jpg"
fnames = [fname for fname in fnames if fname.endswith(end)]
fnames
```

```python
i = 1
for fname in fnames:
    src = os.path.join(fpath, fname)
    dst = "d" + str(i) + '.jpg'
    dst = os.path.join(fpath, dst)
    os.rename(src, dst)
    i += 1
```
